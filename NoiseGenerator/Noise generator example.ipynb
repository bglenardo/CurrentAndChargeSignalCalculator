# Loading noise data
## Current data is from Aldo, Sep 2019

```python
ngen = NoiseGenerator.NoiseGenerator("noise-data/outnoise_tps.csv", peaking_time=1.2)
ngen.load_noise_spectrum()
ngen.plot_noise_spectrum()
```

## Function for sampling the noise distribution

```python
sampling_rate = 2e6 #Hz
buffer_size = 1400e-6 #seconds
n_samples = int(sampling_rate*buffer_size)
for i in range(5):
    t, v = ngen.sample_random_waveform(sampling_rate, n_samples, debug=0)
    fig, ax = plt.subplots(figsize=(12, 8))
    #scale t to microseconds
    t = [_*1e6 for _ in t]
    v = [_*1000 for _ in v] #volts to mV
    ax.plot(t, v)
    ax.set_xlabel("us")
    ax.set_ylabel("mV")
    plt.show()
```
